# Process str.csv

```python
strfpath = "/Users/my/Documents/webull_tick_imags/20250101TSLA/str.csv"
strpd = pd.read_csv(strfpath, header=None)

strpdxx = strpd.drop_duplicates(subset=list(range(1, 10)))
print('removed {} rows duplicates'.format(len(strpd) - len(strpdxx)))

strpdxx = strpdxx[strpdxx[8]!='11/19'] # yahoo has records back till 11/18, 11/19 not collect completed in webull

def _add_intraids(x):
    x = x.sort_values(9)
    x['intraid'] = range(len(x))
    return x
strpdxx = strpdxx.groupby(8).apply(lambda x: _add_intraids(x))
strpdxx.index = range(len(strpdxx))

strpdxx = strpdxx.rename(columns={
    1: 'Open',
    2: 'High',
    3: 'Low',
    4: 'Close',
    7: 'Volume',
})

strpdxx['Date'] = strpdxx[8].apply(lambda x: '2024-'+'-'.join(x.split("/")))
def yf_time(x):
    h, m = x.split(":")
    if m=='00':
        return "{}:{}".format(int(h)-1, '59')
    else:
        return "{}:{}".format(h, int(m)-1)

strpdxx['yftime'] = strpdxx[9].apply(lambda x: yf_time(x))
strpdxx['Datetime'] = strpdxx[['Date', 'yftime']].apply(lambda x: x['Date'] + " " + x['yftime'], axis=1)
```

```python
def webull_download(start='', end='', interval=''):
    webulldb_froot = "/Users/my/Library/Mobile Documents/com~apple~CloudDocs/env/projects/tsla_1yr/webulldb/"
    availfs = os.listdir(webulldb_froot)
    qryfs = [x.strftime("%Y-%m-%d")+"_"+interval+".csv" for x in pd.date_range(start, end)]
    fpds = []
    for f in qryfs:
        if f in availfs:
            fpd = pd.read_csv(webulldb_froot + f)[['Open', 'High', 'Low', 'Close', 'Volume', 'Datetime', 'intraid']]
            fpds.append(fpd.copy(deep=True))
    datapd = pd.concat(fpds).set_index('Datetime')
    datapd.index = pd.to_datetime(datapd.index)
    datapd = datapd.sort_index()
    return datapd
```

```python
def mk_screenshot():
    # make screen shot of ohlcv
    # have adjusted area
    pyautogui.hotkey('shift', 'command', '5')
    pyautogui.hotkey('enter')

def screenshot_to_str(img_path, ):
    return pytesseract.image_to_string(img_path)

def screenshot_to_strfile(img_path, sid='',
        distfroot="/Users/my/Documents/webull_tick_imags/20250101TSLA/", 
        distfname="str.csv"):
    xstr = screenshot_to_str(img_path)
    elements = re.findall("[.0-9%-]+", xstr.replace(",", ""))
    elementstr = sid + "," + ",".join(elements) + '\n'
    with open(distfroot+distfname, 'a') as distf:
        distf.write(elementstr)
    
def imgfolder_to_strfile(
        img_root="/Users/my/Documents/webull_tick_imags/20250101TSLA/ohlcv/", 
        distfroot="/Users/my/Documents/webull_tick_imags/20250101TSLA/", distfname="str.csv", dispose=False):
    fs = [f for f in os.listdir(img_root) if ".png" in f]
    for f in tqdm(fs):
        sid = "".join(
            re.findall("[0-9]", 
            re.findall("[0-9]{4}-[0-9]{2}-[0-9]{2} at [0-9]{2}.[0-9]{2}.[0-9]{2}", f)[0]))
        screenshot_to_strfile(img_root+f, sid, distfroot, distfname)
        if dispose:
            move_from_a_to_b(img_root, distfroot+"bin/", f)
            
            
def move_from_a_to_b(aroot, broot, ffullname):
    import shutil
    shutil.move(aroot+ffullname, broot+ffullname)


# def process_ohlcv_folder(xroot="/Users/my/Documents/webull_tick_imags/20250101TSLA/ohlcv/"):
#     # can do it multi processor; but sc
#     fs = os.listdir(xroot)
```
